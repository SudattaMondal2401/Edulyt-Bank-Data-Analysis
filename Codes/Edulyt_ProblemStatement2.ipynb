# Obtaining the dataframe which contains transactions only for the last 3 months

```python
df6 = df4[df4['date']>=980919] #we can see from df4.describe() that the max value for date is 981219. Thus, we subtract 3 months from it to obtain 980919
```

```python
df6
```

# Adding the condition of most/least population to the previous dataframe

### Dataset containing transactions of the last 3 months for the 5 most populated districts

```python
data1 = df6[df6['district_id']==1]
data2 = df6[df6['district_id']==54] 
data3 = df6[df6['district_id']==74]
data4 = df6[df6['district_id']==70]
data5 = df6[df6['district_id']==68]

data_highest = pd.concat([data1, data2, data3, data4, data5])
```

```python
data_highest['type'].value_counts()
```

```python
data_highest
```

### Dataset containing transactions of the last 3 months for the 5 least populated districts

```python
data1 = df6[df6['district_id']==68]
data2 = df6[df6['district_id']==28] 
data3 = df6[df6['district_id']==12]
data4 = df6[df6['district_id']==30]
data5 = df6[df6['district_id']==18]

data_lowest = pd.concat([data1, data2, data3, data4, data5])
```

```python
data_lowest['type'].value_counts()
```

```python
data_lowest
```

# Credit and Debit calculation for 5 most populated districts in last 3 months

### Credit Amount

```python
cr = data_highest[data_highest['type']=='CREDIT']
print("Total Credit in the last 3 months for the 5 most populated districts is = ", cr['amount'].sum().round())
```

### Debit Amount

```python
db = data_highest[data_highest['type']=='DEBIT']
print("Total Debit in the last 3 months for the 5 most populated districts is = ", db['amount'].sum().round())
```

# Credit and Debit calculation for 5 least populated districts in last 3 months

### Credit Amount

```python
cr = data_lowest[data_lowest['type']=='CREDIT']
print("Total Credit in the last 3 months for the 5 least populated districts is = ", cr['amount'].sum().round())
```

### Debit Amount

```python
db = data_lowest[data_lowest['type']=='DEBIT']
print("Total Debit in the last 3 months for the 5 least populated districts is = ", db['amount'].sum().round())
```
