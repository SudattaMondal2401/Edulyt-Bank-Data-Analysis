# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
df1
```

```python
df2 = df1.drop(['type_x', 'issued', 'type_y'], axis = 1)
```

```python
df2
```

```python
df3 = df2.drop_duplicates(subset = ['account_id'])
```

```python
df3
```

```python
df4 = dataset_loan.drop_duplicates(subset=['account_id'])
```

```python
df5 = df4.drop(['date', 'duration', 'payments', 'status', 'amount'], axis = 1)
```
