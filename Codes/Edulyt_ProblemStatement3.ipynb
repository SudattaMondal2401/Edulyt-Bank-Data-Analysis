# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
df1 = dataset_disposition.merge(dataset_card, how = 'right', on='disp_id') #if on not given, sometimes returns null values for an entire column
```

```python
df1
```

```python
df2 = df1.merge(dataset_client, how = 'left', on='client_id')
```

```python
df2.drop(['Unnamed: 0', 'type_x', 'birth_number', 'district_id', 'disp_id', 'issued'], axis = 1) #This is the final dataframe we will be working with
```

```python
df2.describe()
```
