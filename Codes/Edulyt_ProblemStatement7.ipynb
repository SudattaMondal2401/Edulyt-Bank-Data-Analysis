# Importing the libraries

```python
import pandas as pd
```

# Setting the path and uploading the excel into dataframes

```python
path = "D:\Work\My work\Edulyt Internship\Dataset\Edulyt India - Data for Project - 3.xlsx"
```

```python
dataset_account = pd.read_excel(path, sheet_name = 'New_Account')
dataset_region = pd.read_excel(path, sheet_name = 'District')
dataset_order = pd.read_excel(path, sheet_name = 'Order')
```

```python
dataset_region.rename(columns = {'A1':'district_id', 'A2':'district_name','A3':'region'}, inplace = True)
```

```python
dataset_region.describe()
```

```python
dataset_order.describe()
```

# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
df1 = dataset_account.merge(dataset_region, how = 'left')
```

```python
df1
```

```python
df2 = df1.drop(['frequency', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15','A16'], axis = 1)
df2
```

```python
df3 = dataset_order.merge(df2, how = 'left', on = 'account_id')
df3
```

```python
df3.info()
```

```python
df4 = df3.dropna(subset=['k_symbol'])
```

```python
df4.info()
```

```python
df4['k_symbol'].value_counts()
```

```python
df4
```

# Replacing the k_symbol with proper names

```python
df4['k_symbol'].value_counts()
```

```python
df4.loc[df4['k_symbol'] == 'SIPO', 'k_symbol'] = "household_payment"
df4.loc[df4['k_symbol'] == 'UVER', 'k_symbol'] = "loan_payment"
df4.loc[df4['k_symbol'] == 'POJISTNE', 'k_symbol'] = "insurance_payment"
df4.loc[df4['k_symbol'] == 'LEASING', 'k_symbol'] = "lease_payment"
```

```python
df4['k_symbol'].value_counts()
```

# Finding out the dataframe of each district

```python
df5 = df4.drop(['order_id', 'bank_to', 'account_to'], axis = 1)
df5
```

```python
df4['region'].value_counts()
```

```python
df_sm = df4[df4['region']=='south Moravia']
df_nm = df4[df4['region']=='north Moravia']
df_cb = df4[df4['region']=='central Bohemia']
df_eb = df4[df4['region']=='east Bohemia']
df_wb = df4[df4['region']=='west Bohemia']
df_nb = df4[df4['region']=='north Bohemia']
df_sb = df4[df4['region']=='south Bohemia']
df_p = df4[df4['region']=='Prague']
```
