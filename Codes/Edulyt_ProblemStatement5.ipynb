# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
dataset_region['crimes_committed95'] = pd.to_numeric(dataset_region['crimes_committed95'], errors = 'coerce') 
#Jasenik has a '?' for the no. of crime rates, which gives an error, thus replacing it with null
```

```python
x = dataset_region.loc[:, 'crimes_committed95'].values
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'median')
x_corrected = si.fit_transform(x.reshape(-1, 1))
```

```python
df = pd.DataFrame(x_corrected.reshape(77,1), columns =['crimes_committed95'])
dataset_region['crimes_committed95'] = df
```

```python
dataset_region
```

```python
df1 = dataset_account.merge(dataset_region, how = 'left', on='district_id')
```

```python
df1
```

```python
df2 = df1.drop(['frequency','A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A16'], axis = 1)
```

```python
df2
```

```python
df3 = dataset_loan.merge(df2, how = 'left', on = 'account_id')
```
