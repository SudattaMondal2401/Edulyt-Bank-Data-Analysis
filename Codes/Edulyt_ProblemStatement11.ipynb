# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
df1 = dataset_disposition.merge(dataset_client, how = 'left', on = 'client_id')
```

```python
df1
```

```python
df2 = df1.drop(['Unnamed: 0', 'type', 'birth_number', 'age_levels', 'age'], axis = 1)
```

```python
df2
```

```python
df3 = dataset_card.merge(df2, how = 'left', on = 'disp_id')
```

```python
df3
```

```python
df4 = dataset_region.drop(['A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15', 'A16'], axis = 1)
```

```python
df5 = df3.merge(df4, how = 'left', on = 'district_id')
```

```python
df5
```

# Selecting the customers with GOLD cards

```python
df5['type'].value_counts()
```

```python
df6 = df5[df5['type']=='GOLD']
```

```python
df6
```

# Selecting Male customers with GOLD cards

```python
df7 = df6[df6['gender']=='MALE']
```

```python
df7
```

# Selecting male customers with GOLD cards according to region

###  Male customers with GOLD cards from Bohemia

```python
df_cb = df7[df7['region']=='central Bohemia']
df_eb = df7[df7['region']=='east Bohemia']
df_wb = df7[df7['region']=='west Bohemia']
df_nb = df7[df7['region']=='north Bohemia']
df_sb = df7[df7['region']=='south Bohemia']

df8 = pd.concat([df_cb, df_eb, df_wb, df_nb, df_sb])
```

```python
df8
```

### Male customers with GOLD cards from Moravia

```python
df_sm = df7[df7['region']=='south Moravia']
df_nm = df7[df7['region']=='north Moravia']

df9 = pd.concat([df_sm, df_nm])
```

```python
df9
```

# Printing the results

```python
print('No. of men from Bohemia possessing GOLD cards = ', len(df8))
print('Thus, out of all GOLD cards',  np.round(len(df8)/len(df6) * 100, 3), '% of the cards are issued to men from Bohemia.')
```

```python
print('No. of men from Moravia possessing GOLD cards = ', len(df9))
print('Thus, out of all GOLD cards',  np.round(len(df9)/len(df6) * 100, 3), '% of the cards are issued to men from Moravia.')
```

```python
print("Thus, there are more men in Bohemia with gold cards as compared to men from Moravia. ")
```
