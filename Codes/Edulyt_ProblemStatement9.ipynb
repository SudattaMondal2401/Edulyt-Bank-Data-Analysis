# Importing the libraries

```python
import pandas as pd
from statistics import mean
import matplotlib.pyplot as plt
```

# Setting the path and uploading the excel into dataframes

```python
path = "D:\Work\My work\Edulyt Internship\Dataset\Edulyt India - Data for Project - 3.xlsx"
```

```python
dataset_account = pd.read_excel(path, sheet_name = 'New_Account')
dataset_region = pd.read_excel(path, sheet_name = 'District')
dataset_loan = pd.read_excel(path, sheet_name = 'Loan')
```

# Viewing the datasets and Renaming the required columns

```python
dataset_account.head()
```

```python
dataset_region.head()
```

```python
dataset_loan.head()
```

```python
dataset_account.describe()
```

```python
dataset_region.describe()
```

```python
dataset_region.rename(columns = {'A1':'district_id', 'A2':'district_name','A3':'region', 'A11': 'average_salary', 'A12': 'unemployment_rate95', 'A15': 'crimes_committed95'}, inplace = True)
```

```python
dataset_region.describe()
```

# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
df1 = dataset_account.merge(dataset_region, how = 'left')
```

```python
df1
```

```python
df2 = df1.drop(['frequency', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A13', 'A14', 'A16'], axis = 1)
df2
```

```python
df3 = dataset_loan.merge(df2, how = 'left', on = 'account_id')
df3
```

```python
df3.info()
```

```python
df3['crimes_committed95'] = pd.to_numeric(df3['crimes_committed95'], errors = 'coerce') 
```

```python
df3.info()
```

```python
df3['unemployment_rate95'] = pd.to_numeric(df3['unemployment_rate95'], errors = 'coerce')
```

```python
df3.info()
```

```python
x = df3.loc[:, 'unemployment_rate95'].values
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'median')
x_corrected = si.fit_transform(x.reshape(-1, 1))
```

```python
df = pd.DataFrame(x_corrected.reshape(682,1), columns =['unemployment_rate95'])
df3['unemployment_rate95'] = df
df3.info()
```

```python
x = df3.loc[:, 'crimes_committed95'].values
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = 'median')
x_corrected = si.fit_transform(x.reshape(-1, 1))
```

```python
df = pd.DataFrame(x_corrected.reshape(682,1), columns =['crimes_committed95'])
df3['crimes_committed95'] = df
df3.info()
```

```python
df3
```

# Finding out the dataframe of each district

```python
df4 = df3.drop(['loan_id', 'date', 'duration', 'payments'], axis = 1)
df4
```

```python
df4['region'].value_counts()
```

```python
df_sm = df4[df4['region']=='south Moravia']
df_nm = df4[df4['region']=='north Moravia']
df_cb = df4[df4['region']=='central Bohemia']
df_eb = df4[df4['region']=='east Bohemia']
df_wb = df4[df4['region']=='west Bohemia']
df_nb = df4[df4['region']=='north Bohemia']
df_sb = df4[df4['region']=='south Bohemia']
df_p = df4[df4['region']=='Prague']
```

### South Moravia

```python
print('The loan payment status according to each category is:')
print(df_sm['status'].value_counts())
```

```python
d = df_sm[df_sm['status']=='A']
print('Highest amount for type A loan payment = ', max(d['amount']))
a = max(d['amount'])
for i, j in zip(d['amount'], d['account_id']):
    if (max(d['amount'])==i):
        print('The account id with the highest amount for type A loan payment is ',j)
print()        
d = df_sm[df_sm['status']=='B']
print('Highest amount for type B loan payment = ', max(d['amount']))
b = max(d['amount'])
for i, j in zip(d['amount'], d['account_id']):
    if (max(d['amount'])==i):
        print('The account id with the highest amount for type B loan payment is ',j)
print()       
d = df_sm[df_sm['status']=='C']
c = max(d['amount'])
print('Highest amount for type C loan payment = ', max(d['amount']))
for i, j in zip(d['amount'], d['account_id']):
    if (max(d['amount'])==i):
        print('The account id with the highest amount for type C loan payment is ',j)
print()       
d = df_sm[df_sm['status']=='D']
d = max(d['amount'])
print('Highest amount for type D loan payment = ', max(d['amount']))
for i, j in zip(d['amount'], d['account_id']):
    if (max(d['amount'])==i):
        print('The account id with the highest amount for type D loan payment is ',j)
print()
```

```python
print('The average mean-salary for the region =', round(mean(df_sm['average_salary']), 3))
print('The average unemployment rate for 1995 for the region =', round(mean(df_sm['unemployment_rate95']), 3))
print('The average number of crime committed for 1995 for the region =', round(mean(df_sm['crimes_committed95']), 3))
```

```python
label.append('South Moravia')
l_loan.append(a+c)
l_debt.append(b+d)
salary.append(round(mean(df_sm['average_salary']), 3))
unemployment.append(round(mean(df_sm['unemployment_rate95']), 3))
crime.append(round(mean(df_sm['crimes_committed95']), 3))
```

# Relating the loans and debt of each region to crimes, unemployment and average salary graphically

```python
print(label)
print(l_loan)
print(l_debt)
print(salary)
print(unemployment)
print(crime)
```

```python
plt.plot(l_loan, salary, c = 'green')
plt.scatter(l_loan, salary, c = 'black')
for i, j, k in zip(l_loan, salary, label):
    plt.text(i, j, k, va='bottom', ha='center')
    
plt.xlabel('Loans repaid')
plt.ylabel('Average Salary')
plt.title('Loans repaid or being repaid vs Average Salary')
plt.show()
```

```python
plt.plot(l_debt, salary, c = 'red')
plt.scatter(l_debt, salary, c = 'black')
for i, j, k in zip(l_debt, salary, label):
    plt.text(i, j, k, va='bottom', ha='center')
    
plt.xlabel('Loans in debt')
plt.ylabel('Average Salary')
plt.title('Loans in debt vs Average Salary')
plt.show()
```

```python
plt.plot(l_loan, crime, c = 'green')
plt.scatter(l_loan, crime, c = 'black')
for i, j, k in zip(l_loan, crime, label):
    plt.text(i, j, k, va='bottom', ha='center')
    
plt.xlabel('Loans repaid')
plt.ylabel('Average Crime Rate 1995')
plt.title('Loans repaid or being repaid vs Crime Rate')
plt.show()
```

```python
plt.plot(l_debt, crime, c = 'red')
plt.scatter(l_debt, crime, c = 'black')
for i, j, k in zip(l_debt, crime, label):
    plt.text(i, j, k, va='bottom', ha='center')
    
plt.xlabel('Loans in debt')
plt.ylabel('Average Crime Rate 1995')
plt.title('Loans in debt vs Crime Rate')
plt.show()
```

```python
plt.plot(l_loan, unemployment, c = 'green')
plt.scatter(l_loan, unemployment, c = 'black')
for i, j, k in zip(l_loan, unemployment, label):
    plt.text(i, j, k, va='bottom', ha='center')
    
plt.xlabel('Loans Repaid')
plt.ylabel('Average Unemployment Rate 1995')
plt.title('Loans repaid or being repaid vs Unemployment Rate')
plt.show()
```
