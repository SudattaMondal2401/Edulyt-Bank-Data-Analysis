# Segregrating the data of the last 3 months month-wise

```python
df6 = df5[df5['date']>=980919] #we can see from df4.describe() that the max value for date is 981219. Thus, we subtract 3 months from it to obtain 980919
```

```python
df6
```

```python
month = []
for i in df6['date']:
    if (i>=980919 and i<981019):
        month.append('19-9-1998 to 18-10-1998')
    elif (i>=981019 and i<981119):
        month.append('19-10-1998 to 18-11-1998')
    elif (i>=981119 and i<=981219):
        month.append('19-11-1998 to 19-12-1998')
        
print(month)
```

```python
df6['months'] = month
```
