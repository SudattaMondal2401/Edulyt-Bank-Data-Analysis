# Merging the dataframes and dropping redundant columns in order to obtain the final dataframe

```python
dataset_region.drop(['A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10', 'A12', 'A5', 'A13', 'A14', 'A15', 'A16'], axis = 1, inplace = True)
```

```python
df2 = dataset_card.merge(dataset_disposition, how = 'left', on = 'disp_id')
```

# Finding out the districts with average salary > 9000

```python
df4 = df3.copy()
df5 = df4[df4['Avg_Salary']>=9000]
```

```python
df5
```

# Printing the results

```python
print("Total number of cards issued = ", len(df3))
print("No. of cards issued for regions where average salary is greater than or equal to 9000 = ", len(df5))
print("% of cards from regions where average salary is greater than or equal to 9000 = ",  len(df5)/len(df3) * 100)
```

### Conclusion - Thus, only 48.99% of the cards have been issued from regions where the average salary is greater than or equal to 9000, which is less than 50% of all the cards issued. Thus, it is not a good strategy to issue cards only in those regions.
